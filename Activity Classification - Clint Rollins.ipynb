```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
from sklearn import tree


training_data = pd.read_csv('C:/Users/rolli/Desktop/WinPython-64bit-3.5.2.3/notebooks/Testing/training.csv',
                            sep= ',', header= None)
test_data = pd.read_csv('C:/Users/rolli/Desktop/WinPython-64bit-3.5.2.3/notebooks/Testing/testing.csv',
                            sep= ',', header= None)

X = training_data.values[:,1:4]
Y = training_data.values[:,0]

X_test = test_data.values[:,1:4]
Y_test = test_data.values[:,0]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)

pred = clf.predict(X_test)

df = pd.DataFrame(np.column_stack((Y_test,pred)), columns=['Actual', 'Predicted'])
print("\nAccuracy for accelerometer only with mean feature is ", accuracy_score(Y_test,pred))

df
```
